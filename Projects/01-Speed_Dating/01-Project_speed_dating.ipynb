```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('Speed_Dating_Data.csv', encoding = 'ISO-8859-1')
dataset
```

```python
pd.set_option('display.max_columns', None)
dataset
```

```python
diid = dataset.drop_duplicates(subset = "iid")
```

```python
diid
```

```python
diid = diid.rename(columns={'gender': 'sexe'})

diid['sexe'] = diid['sexe'].apply(lambda x:
                            "Femme" if x == 0
                            else "Homme" if x == 1 else x)

diid.head()
```

```python
sns.countplot(diid['sexe'])
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.distplot(diid['age'])
```

```python
dataset1 = dataset[['iid','gender','wave','pid','match','int_corr','samerace','age_o','race_o','dec_o','attr_o', 
                    'sinc_o','intel_o','fun_o','amb_o','shar_o','age','field','race','imprace','imprelig', 
                    'income','career','exphappy','expnum']]
dataset1.head()
```

```python
dataset1 = dataset1.rename(columns={'iid': 'n° candidat', 'gender': 'sexe', 'wave': 'vague', 'pid': 'n° partenaire', 
                                    'int_corr': 'intérêt_commun', 'samerace': 'même_ethnie', 'age_o': 'age_partenaire',
                                   'race_o': 'ethnie_partenaire','dec_o': 'décision_partenaire', 'attr_o': 'attractivité_partenaire', 
                                    'sinc_o': 'sincérité_partenaire','intel_o': 'intelligence_partenaire', 'fun_o': 'fun_partenaire', 
                                    'amb_o': 'ambition_partenaire','shar_o': 'désirs communs', 'field': 'domaine_études', 
                                    'race': 'ethnie_candidat', 'imprace': 'importance_ethnie','imprelig': 'importance_religion', 
                                    'income': 'revenus', 'career': 'domaine_activité', 'exphappy': 'joie_participé', 
                                    'expnum': 'espérance_de _plaire'
                                   })
dataset1.head()
```

```python
dataset1['sexe'] = dataset1['sexe'].apply(lambda x:
                            "Femme" if x == 0
                            else "Homme" if x == 1 else x)

dataset1.head()
```

```python
dataset1['match'] = dataset1['match'].apply(lambda x :
                               "Non" if x==0
                               else "Oui" if x==1 else x)

dataset1.head()
```

```python
dataset1['même_ethnie'] = dataset1['même_ethnie'].apply(lambda x :
                                                        "Non" if x==0
                                                        else "Oui" if x==1 else x) 

dataset1.head()
```

```python
dataset1['décision_partenaire'] = dataset1['décision_partenaire'].apply(lambda x :
                                                                    "Non" if x==0
                                                                    else "Oui" if x==1 else x)

dataset1.head()
```

```python
dataset1['ethnie_partenaire'] = dataset1['ethnie_partenaire'].apply(lambda x :
                                             "Noir américain" if x==1
                                            else "Caucasian" if x==2
                                            else "Latino" if x==3
                                            else "Asiatique" if x==4
                                            else "Amérindien" if x==5
                                            else "Autre" if x==6 else x)

dataset1.head()
```

```python
dataset1['ethnie_candidat'] = dataset1['ethnie_candidat'].apply(lambda x :
                                             "Noir américain" if x==1
                                            else "Caucasian" if x==2
                                            else "Latino" if x==3
                                            else "Asiatique" if x==4
                                            else "Amérindien" if x==5
                                            else "Autre" if x==6 else x)

dataset1.head()
```
